```python
#导入第三方包
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import scipy.stats.mstats as mst

import matplotlib as mpl


plt.style.use('ggplot')

#设定目录及读取数据文件
dir = './data/prophet/'
brand1 = pd.read_table(dir + 'tbrand1.csv',engine='python',sep=',')
brand2 = pd.read_table(dir + 'tbrand2.csv',engine='python',sep=',')
brand3 = pd.read_table(dir + 'tbrand3.csv',engine='python',sep=',')
brand4 = pd.read_table(dir + 'tbrand4.csv',engine='python',sep=',')
brand5 = pd.read_table(dir + 'tbrand5.csv',engine='python',sep=',')
answer=pd.read_table(dir + 'answer_A_20180225.txt',engine='python',header=None)
answer.columns = ['date','cnt']
```

```python
from sklearn.externals import joblib
data1=brand1[['day_of_week','date']]
data2=brand2[['day_of_week','date']]
data3=brand3[['day_of_week','date']]
data4=brand4[['day_of_week','date']]
data5=brand5[['day_of_week','date']]
```

```python
model=['LR_brand','RF_brand', 'DT_brand','ADA_brand','XGB_brand']
for i in model:
    model = joblib.load(dir+i+'1.pkl')
    brand1[i]=model.predict(data1)
print(brand1.head())

model=['LR_brand','RF_brand', 'DT_brand','ADA_brand','XGB_brand']
for i in model:
    model = joblib.load(dir+i+'2.pkl')
    brand2[i]=model.predict(data2)
print(brand2.head())

model=['LR_brand','RF_brand', 'DT_brand','ADA_brand','XGB_brand']
for i in model:
    model = joblib.load(dir+i+'3.pkl')
    brand3[i]=model.predict(data3)
print(brand3.head())

model=['LR_brand','RF_brand', 'DT_brand','ADA_brand','XGB_brand']
for i in model:
    model = joblib.load(dir+i+'4.pkl')
    brand4[i]=model.predict(data4)
print(brand4.head())

model=['LR_brand','RF_brand', 'DT_brand','ADA_brand','XGB_brand']
for i in model:
    model = joblib.load(dir+i+'5.pkl')
    brand5[i]=model.predict(data5)
print(brand5.head())

    
```

```python
frames=[brand1,brand2,brand3,brand4,brand5]
df=pd.concat(frames)
df=df.sort_values(by='date')
df.head(20)
```

```python
from sklearn.externals import joblib
modelx1=joblib.load(dir+'x1.pkl')
testdata=df[['date', 'day_of_week', 'brand', 'LR_brand', 'RF_brand', 'DT_brand', 'ADA_brand', 'XGB_brand']]
df['x1']=modelx1.predict(testdata)
```

```python
df2=df
del df2['brand']
df2=df2.groupby(['date','day_of_week'],as_index=False).sum()
df2.head(20)
```
